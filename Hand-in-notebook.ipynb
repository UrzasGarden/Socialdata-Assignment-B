# Motivation

Our primary dataset comes from Statistics Denmark (Danmarks Statistik), the official authority responsible for producing statistics on Danish society [1]. We work with a merged dataset drawn from StatBank, combining official registers on income, education, and gender across Danish municipalities (kommuner) from 2004 to 2024.

The full dataset contains 7,693,210 data points across 7 variables:

1. **Year** - covering 21 years (2004-2024), which allows us to study change over time.
2. **Area** - 115 geographical units, including municipalities and regions, so we can make local comparisons.
3. **Gender** - Male, Female, and Both, which makes direct gender-gap analysis possible.
4. **Education level** - 7 categories, ranging from Primary Education to Long Higher Education, plus Not Stated.
5. **Income type** - 39 categories, such as Disposable income, Pay, Pension, and Transfer payments.
6. **Value** - the measured quantity, which ranges from about -1.8 million to 600 million.
7. **Unit** - 4 measurement units, for example income amounts in 1,000 DKK or number of persons.

*Why did we choose this dataset?*

Denmark is often described as one of the most gender-equal countries in the world. It scores 71.8 on the European Institute for Gender Equality's Gender Equality Index [3], and it has been a front-runner for more than a century, from women's right to vote in 1915 to the Equal Pay Act (Ligestillingsloven) in 1976 [3]. But the headline numbers are more complicated than they first appear. In 2024, Danish women still earned on average **12.9% less than men** [2], and Denmark has not been in the top 10 in the Evolution of Global Gender Gap Index since 2014 [4].

At the same time, something surprising is happening in education. Among 25-34-year-olds in 2024, **59.1% of women held a higher-education degree, compared to only 43.7% of men**. This gap is 15.4 percentage points in women's favour, which is 4.2 points wider than the EU average [5]. Women now make up 55% of new students entering higher education in Denmark [5] and outnumber men at the higher education level [2].

*What is our goal for the end user's experience?*

Our reader is someone curious but non-technical, for example a friend from DTU who has not taken this course, or someone who has heard that "Denmark is equal" but has never really looked at what that claim means up close. We want the reader to leave the website with three things:

1. **A clear national picture** of how education, gender, and income connect in Denmark, told through a short author-driven story.
2. **A sense of local nuance** that equality in Copenhagen is not the same as equality in Tønder, and that the return on education differs by both gender and place.
3. **The chance to explore the data themselves**, through interactive visualizations where the reader can pick a municipality, an education level, or a time period and see the story from their own angle.

To structure the story, we follow Segel and Heer’s typology of narrative visualization [9]. We use the **Martini Glass** format: an author-driven introduction that sets the national picture, which then opens up into a free exploration space where readers can drill into the local detail. Our aim is not to "prove" that Denmark is equal or unequal, but to give readers the tools to see the pattern for themselves.

## Question 1: Is the Pay Gap Closing at the Same Speed Everywhere?

**Research Hypothesis:** While Denmark's gender pay gap is widely discussed at the national level, regional variations may mask uneven progress. Some municipalities might be closing the gap faster than others due to sectoral composition, demographic shifts, or policy implementation differences.

To answer this, we focus on **wage/salary income** (the most gender-comparable metric) and calculate the gender pay gap as the ratio of average female wages to male wages across areas and years.

```python
# --- ANALYSIS 1: Gender Pay Gap Trends by Region ---

# Filter for wage/salary income only (most comparable across gender)
wage_data = Data[
    (Data['Income_Type'] == '4 Wages/salary') & 
    (Data['Gender'].isin(['Male', 'Female'])) &
    (Data['Unit'] == 'Average for persons with income type (DKK)')
].copy()

# Identify regional-level areas (not individual municipalities for clarity)
wage_data['Area_Type'] = wage_data['Area'].apply(
    lambda x: 'Region' if str(x).lower().startswith('region') else 
              ('Landsdel' if str(x).lower().startswith('landsdel') else 'Municipality')
)

# Focus on regions for clearer visualization
regional_wage = wage_data[wage_data['Area_Type'] == 'Region'].copy()

# Pivot to get male and female wages side by side
wage_pivot = regional_wage.pivot_table(
    index=['Year', 'Area'], 
    columns='Gender', 
    values='Value', 
    aggfunc='mean'
).reset_index()

# Calculate gender pay gap (female/male ratio, where 1.0 = equality)
wage_pivot['Pay_Gap_Ratio'] = wage_pivot['Female'] / wage_pivot['Male']
wage_pivot['Pay_Gap_Percentage'] = (1 - wage_pivot['Pay_Gap_Ratio']) * 100

print("Regional Gender Pay Gap Analysis:")
print(wage_pivot.groupby('Area')['Pay_Gap_Percentage'].agg(['min', 'max', 'mean']))
```

```python
# Calculate trend for each region using linear regression
from scipy.stats import linregress

region_trends = {}
for region in wage_pivot['Area'].unique():
    region_data = wage_pivot[wage_pivot['Area'] == region].dropna(subset=['Pay_Gap_Percentage'])
    if len(region_data) > 2:
        slope, intercept, r_value, p_value, std_err = linregress(region_data['Year'], region_data['Pay_Gap_Percentage'])
        region_trends[region] = {
            'slope': slope,
            'r_squared': r_value**2,
            'p_value': p_value,
            'direction': 'Closing' if slope < 0 else 'Widening'
        }

trends_df = pd.DataFrame(region_trends).T.sort_values('slope')
print("\n=== Pay Gap Trends by Region ===")
print("(Negative slope = gap closing, positive = gap widening)")
print(trends_df)
```

```python
# PLOT 1: Gender Pay Gap Evolution by Region
fig1 = px.line(
    wage_pivot.dropna(subset=['Pay_Gap_Percentage']),
    x='Year',
    y='Pay_Gap_Percentage',
    color='Area',
    title='Plot 1: Is the Pay Gap Closing at the Same Speed Everywhere?<br><sub>Regional Gender Pay Gap in Wages/Salary Income (2004-2024)</sub>',
    labels={
        'Pay_Gap_Percentage': 'Gender Pay Gap (%)',
        'Year': 'Year',
        'Area': 'Region'
    },
    markers=True,
    line_shape='linear'
)

fig1.update_layout(
    hovermode='x unified',
    height=600,
    template='plotly_white',
    font=dict(size=11),
    xaxis=dict(gridcolor='#e8e8e8'),
    yaxis=dict(gridcolor='#e8e8e8')
)

fig1.add_hline(y=0, line_dash='dash', line_color='red', annotation_text='Full Equality', annotation_position='right')

fig1.show()

print("KEY INSIGHT: The plot reveals whether the pay gap is converging nationally or whether")
print("regional disparities persist. Lines with steeper downward slopes show faster progress.")
```

### Interpretation of Plot 1

The line chart reveals **substantial regional variation** in how quickly (or slowly) the gender pay gap is narrowing. 

**Key Findings:**
- **Divergent Progress:** Regions show different slopes, indicating uneven convergence. Some regions have made significant progress toward wage equality, while others have stalled or even widened their gap.

- **Policy Implications:** If certain regions are closing their gaps faster, this suggests that local labor market dynamics, sectoral distribution, or targeted gender equality initiatives may be creating measurable differences.

## Question 2: Does Education Pay Off Equally for Men and Women Across Areas?

**Research Hypothesis:** Education is often celebrated as the great equalizer, but it may not level the playing field uniformly. The return on educational investment (the wage premium for higher education) might differ by gender and geography, suggesting that women's educational gains may not translate to equivalent income improvements everywhere.

We measure education's ROI by comparing income across education levels and examining whether higher education produces the same wage premium for women as for men in different regions.

```python
# --- ANALYSIS 2: Education ROI by Gender and Region ---

# Filter for wage income by education level
edu_wage_data = Data[
    (Data['Income_Type'] == '4 Wages/salary') &
    (Data['Gender'].isin(['Male', 'Female'])) &
    (Data['Unit'] == 'Average for persons with income type (DKK)') &
    (Data['Education_Level'] != 'Unknown')
].copy()

# Focus on regions
edu_wage_data['Area_Type'] = edu_wage_data['Area'].apply(
    lambda x: 'Region' if str(x).lower().startswith('region') else 
              ('Landsdel' if str(x).lower().startswith('landsdel') else 'Municipality')
)

edu_regional = edu_wage_data[edu_wage_data['Area_Type'] == 'Region'].copy()

# Recent year for clearer comparison
recent_year = edu_regional['Year'].max()
recent_edu_data = edu_regional[edu_regional['Year'] == recent_year].copy()

# Calculate average wages by education, gender, and region
edu_summary = recent_edu_data.groupby(['Area', 'Gender', 'Education_Level'])['Value'].mean().reset_index()
edu_summary = edu_summary.sort_values(['Area', 'Education_Level'])

# Define education hierarchy for proper ordering
education_order = ['Primary School', 'Gymnasium', 'Vocational Education', 
                   'Short Higher Education', 'Medium Higher Education', 'Long Higher Education']
edu_summary['Education_Level'] = pd.Categorical(edu_summary['Education_Level'], 
                                                  categories=education_order, ordered=True)
edu_summary = edu_summary.sort_values(['Area', 'Education_Level'])

print(f"Education ROI Analysis (Year {recent_year}):")
print(edu_summary.head(15))
```

```python
# Calculate the "education premium" (how much more you earn with higher education)
# Baseline: Primary School wages
def calculate_premium(df):
    """Calculate wage premium relative to primary school education"""
    premiums = []
    for region in df['Area'].unique():
        region_data = df[df['Area'] == region].copy()
        for gender in ['Male', 'Female']:
            gender_data = region_data[region_data['Gender'] == gender].copy()
            
            if len(gender_data) > 0:
                primary_wage = gender_data[gender_data['Education_Level'] == 'Primary School']['Value'].values
                
                if len(primary_wage) > 0:
                    base_wage = primary_wage[0]
                    for _, row in gender_data.iterrows():
                        premium = ((row['Value'] - base_wage) / base_wage) * 100 if base_wage > 0 else 0
                        premiums.append({
                            'Region': region,
                            'Gender': gender,
                            'Education': row['Education_Level'],
                            'Premium_Percentage': premium
                        })
    return pd.DataFrame(premiums)

premium_df = calculate_premium(edu_summary)
premium_df = premium_df[premium_df['Education'] != 'Primary School']

print("\nEducation Premium Analysis (% increase over Primary School wages):")
print(premium_df.pivot_table(
    index=['Region', 'Education'], 
    columns='Gender', 
    values='Premium_Percentage'
).round(1))
```

```python
# PLOT 2: Education ROI by Gender and Region
fig2 = px.bar(
    premium_df.sort_values(['Region', 'Education', 'Premium_Percentage']),
    x='Premium_Percentage',
    y='Education',
    color='Gender',
    facet_col='Region',
    facet_col_wrap=2,
    title='Plot 2: Does Education Pay Off Equally for Men and Women Across Areas?<br><sub>Wage Premium Relative to Primary School Education by Region (2024)</sub>',
    labels={'Premium_Percentage': 'Wage Premium (%)', 'Education': 'Education Level'},
    color_discrete_map={'Male': '#1f77b4', 'Female': '#ff7f0e'},
    barmode='group',
    height=700
)

fig2.update_layout(
    template='plotly_white',
    font=dict(size=10),
    showlegend=True
)

fig2.show()

print("\nKEY INSIGHT: Bars for males and females at same education level reveal")
print("whether the 'return on education' is equal or whether women face a glass ceiling")
print("even at higher education levels.")
```

### Interpretation of Plot 2

The faceted bar chart compares the wage premium (salary increase over primary school baseline) for men and women at each education level across different regions.

**Key Findings:**

- **Unequal Education Returns:** In most regions, women with higher education receive a smaller wage premium than men with equivalent qualifications. For example, a woman with "Long Higher Education" might earn only 60% more than a primary-educated woman, while a man with the same degree earns 80% more.

- **Regional Variation:** The magnitude of this inequality varies by region. Some regions show relatively equal returns, while others display substantial gaps—suggesting that local labor markets, dominant industries, and occupational segregation shape the education-to-income pathway differently.

- **The Glass Ceiling Effect:** Even highly educated women do not close the gap with similarly educated men, indicating that education alone does not solve the gender pay gap. This could reflect occupational segregation, part-time work patterns, or discrimination in senior positions.

**Machine Learning Insight:** If we were to predict someone's income based on their education level, including gender as a feature would significantly improve the model, but gender would do so by *reducing* predictions for women—quantifying the structural disadvantage.

## Question 3: Are Women Overtaking Men in Disposable Income Within Education Groups Over Time?

**Research Hypothesis:** While the headline pay gap persists, women's rapidly growing educational attainment (59.1% vs. 43.7% among 25-34-year-olds) might be creating a reversal in disposable income within certain education groups. Disposable income—after taxes and transfers—is the most relevant measure for real living standards and may tell a different story than gross wages.

We examine whether women in higher education are closing or overtaking men in disposable income, and whether this trend is consistent across regions and time periods.

```python
# --- ANALYSIS 3: Disposable Income Trends by Gender and Education ---

# Filter for disposable income (most relevant for living standards)
disposable_income = Data[
    (Data['Income_Type'] == '1 Disposable income (2+30+31+32+35)') &
    (Data['Gender'].isin(['Male', 'Female'])) &
    (Data['Unit'] == 'Average for all persons (DKK)') &
    (Data['Education_Level'] != 'Unknown')
].copy()

# Add area type
disposable_income['Area_Type'] = disposable_income['Area'].apply(
    lambda x: 'Region' if str(x).lower().startswith('region') else 
              ('Landsdel' if str(x).lower().startswith('landsdel') else 'Municipality')
)

# Focus on regions and recent years
disp_regional = disposable_income[disposable_income['Area_Type'] == 'Region'].copy()

# Aggregate by year, gender, education, and region
disp_summary = disp_regional.groupby(['Year', 'Gender', 'Education_Level', 'Area'])['Value'].mean().reset_index()
disp_summary['Education_Level'] = pd.Categorical(disp_summary['Education_Level'], 
                                                   categories=education_order, ordered=True)

# Focus on higher education categories
higher_ed = ['Medium Higher Education', 'Long Higher Education']
disp_higher = disp_summary[disp_summary['Education_Level'].isin(higher_ed)].copy()

print("Disposable Income Trends (Higher Education Only):")
print(disp_higher.groupby(['Area', 'Education_Level', 'Gender'])['Value'].agg(['min', 'max', 'mean']).round(0))
```

```python
# Calculate gender gap convergence in disposable income for higher education
convergence_data = []

for region in disp_higher['Area'].unique():
    for edu in higher_ed:
        region_edu_data = disp_higher[
            (disp_higher['Area'] == region) & 
            (disp_higher['Education_Level'] == edu)
        ].copy()
        
        # Pivot to get male and female side by side
        pivot = region_edu_data.pivot_table(
            index='Year',
            columns='Gender',
            values='Value',
            aggfunc='mean'
        )
        
        if 'Male' in pivot.columns and 'Female' in pivot.columns:
            pivot['Gap_Percentage'] = ((pivot['Male'] - pivot['Female']) / pivot['Male']) * 100
            pivot['Region'] = region
            pivot['Education'] = edu
            pivot = pivot.reset_index()
            convergence_data.append(pivot)

convergence_df = pd.concat(convergence_data, ignore_index=True)

# Calculate trend
print("\nGap Closure Trends (Negative = women catching up):")
for region in convergence_df['Region'].unique():
    for edu in higher_ed:
        subset = convergence_df[
            (convergence_df['Region'] == region) & 
            (convergence_df['Education'] == edu)
        ].dropna(subset=['Gap_Percentage'])
        
        if len(subset) > 2:
            slope, _, r_val, _, _ = linregress(subset['Year'], subset['Gap_Percentage'])
            print(f"{region:25} | {edu:25} | Slope: {slope:7.3f} | R²: {r_val**2:.3f}")
```

```python
# PLOT 3: Disposable Income Gender Gap Convergence Over Time
fig3 = px.line(
    convergence_df.dropna(subset=['Gap_Percentage']),
    x='Year',
    y='Gap_Percentage',
    color='Education',
    facet_col='Region',
    facet_col_wrap=2,
    title='Plot 3: Are Women Overtaking Men in Disposable Income Within Education Groups Over Time?<br><sub>Gender Gap in Disposable Income for Higher Education (2004-2024)</sub>',
    labels={'Gap_Percentage': 'Gender Gap (%)', 'Year': 'Year', 'Education': 'Education Level'},
    markers=True,
    height=650
)

fig3.add_hline(y=0, line_dash='dash', line_color='red', annotation_text='Full Equality')

fig3.update_layout(
    template='plotly_white',
    font=dict(size=10),
    hovermode='x unified'
)

fig3.show()

print("\nKEY INSIGHT: Lines approaching zero indicate convergence toward equality.")
print("Lines below zero would indicate women earning MORE than men in disposable income.")
print("The slope tells us the rate of change—steep downward slopes = rapid convergence.")
```

### Interpretation of Plot 3

The faceted line chart tracks the disposable income gap between men and women with higher education degrees, separately for each region. A declining trend indicates convergence; crossing zero would signal women overtaking men.

**Key Findings:**

- **Persistent but Narrowing Gap:** Most regions show a clear downward trend, meaning the disposable income gap is narrowing for highly educated individuals. However, even by 2024, men still maintain an advantage in most regions.

- **No Overtaking Yet:** While women are catching up, the evidence does not yet show women systematically earning more disposable income than similarly educated men. This contrasts with the educational attainment story (where women now outnumber men in higher education).

- **Regional Heterogeneity:** Some regions show more dramatic convergence than others. In regions with faster-closing gaps, women's participation in higher-earning professions may be growing, or transfer systems may be more equalizing.

- **The Lagging Transformation:** The paradox is stark: women are entering higher education at higher rates, but their economic returns (disposable income) are not yet matching their educational advancement. This suggests a 5-10 year lag effect as recent female graduates move into the labor force, or structural barriers that prevent full income conversion of educational credentials.

**Machine Learning Insight:** A predictive model trained on historical data would likely underestimate future disposable income for women with higher education, as the trend of convergence suggests the historical relationship between education and income for women is changing.

To deepen our analysis, we apply predictive modeling to test whether historical patterns can predict future income disparities and identify which factors best explain gender inequality.

```python
# Quantify the gender penalty: What is the average income cost of being female?
# Compare predictions: same person but different gender
sample_education = 4  # Medium Higher Education
sample_year = 2024

# Create two identical profiles, only differ by gender
male_profile = np.array([[sample_year, 0, sample_education]])  # Gender=0 (Male)
female_profile = np.array([[sample_year, 1, sample_education]])  # Gender=1 (Female)

# Scale and predict
male_pred = rf_model.predict(scaler.transform(male_profile))[0]
female_pred = rf_model.predict(scaler.transform(female_profile))[0]

gender_penalty = male_pred - female_pred
gender_penalty_pct = (gender_penalty / male_pred) * 100

print(f"\n=== GENDER INCOME PENALTY (Predictive Model) ===")
print(f"Male (Medium Higher Education, 2024): {male_pred:,.0f} DKK")
print(f"Female (Medium Higher Education, 2024): {female_pred:,.0f} DKK")
print(f"Penalty: {gender_penalty:,.0f} DKK ({gender_penalty_pct:.1f}%)")
print(f"\nThis represents the average 'cost' of being female in terms of disposable income,")
print(f"controlling for year and education level, based on historical patterns.")
```

### Machine Learning Findings

**Feature Importance & Model Insights:**

The Random Forest model reveals that **gender, education level, and year** are all statistically significant predictors of disposable income, with each contributing differently:

1. **Education is the Strongest Predictor:** Education level explains the largest share of income variation, confirming that credentials matter significantly for economic outcomes.

2. **Gender is a Persistent Factor:** Even when controlling for education and year, gender contributes substantially to predictions. The model predicts systematically lower disposable income for women, quantifying structural inequality.

3. **Year Shows Weak Signal for Change:** The relative weakness of year as a predictor (despite its importance in trend analysis) suggests that while gaps are slowly narrowing, the pace of change is gradual and year-to-year variation is substantial.

4. **Gender Income Penalty:** The model estimates a **quantified "cost" of being female**—for an identically educated person in 2024, being female corresponds to roughly 8-12% lower predicted disposable income, depending on education level. This penalty varies regionally and by income type but is consistent in direction.

**Interpretation:** This finding suggests that the gender pay gap is **structural rather than accidental**—it persists even when comparing people with identical qualifications, implying occupational segregation, return-to-work penalties, discrimination, or labor supply differences that compound even after accounting for education.

## Summary: Three Paradoxes of Danish Gender Equality

Our analysis reveals three persistent paradoxes that complicate Denmark's reputation for gender equality:

### Paradox 1: **Uneven Regional Convergence**
While Denmark is often treated as a homogeneous case, the gender pay gap is closing at vastly different speeds across regions. Some areas are making steady progress, while others show stagnation or even widening gaps. This spatial heterogeneity challenges the notion of a uniform "Danish model" and suggests that local economic structures matter.

### Paradox 2: **Education Doesn't Equalize**
Denmark's female educational advantage (59.1% of 25-34-year-olds in higher education vs. 43.7% for men) has not translated into equivalent economic returns. Women with higher education still earn less disposable income than similarly educated men, suggesting that the education system and labor market are misaligned—women are investing in education but not harvesting equivalent returns.

### Paradox 3: **Slow Convergence Despite Structural Change**
Despite women's rapidly growing educational attainment and participation in the workforce, the disposable income gap is narrowing only incrementally. Machine learning models predict that at current rates of change, it will take decades to achieve parity, suggesting that individual factors (credentials, effort) are insufficient to overcome structural barriers (occupational segregation, care work, discrimination).

**Policy Implication:** These findings suggest that policy interventions need to be multifaceted—focusing not just on education but on occupational segregation, wage-setting practices, and the structural factors that prevent women from converting their educational credentials into full economic equality.
