# Motivation

Our primary dataset comes from Statistics Denmark (Danmarks Statistik), the official authority responsible for producing statistics on Danish society [1]. We work with a merged dataset drawn from StatBank, combining official registers on income, education, and gender across Danish municipalities (kommuner) from 2004 to 2024.

The full dataset contains 7,693,210 data points across 7 variables:

1. **Year** - covering 21 years (2004-2024), which allows us to study change over time.
2. **Area** - 115 geographical units, including municipalities and regions, so we can make local comparisons.
3. **Gender** - Male, Female, and Both, which makes direct gender-gap analysis possible.
4. **Education level** - 7 categories, ranging from Primary Education to Long Higher Education, plus Not Stated.
5. **Income type** - 39 categories, such as Disposable income, Pay, Pension, and Transfer payments.
6. **Value** - the measured quantity, which ranges from about -1.8 million to 600 million.
7. **Unit** - 4 measurement units, for example income amounts in 1,000 DKK or number of persons.

*Why did we choose this dataset?*

Denmark is often described as one of the most gender-equal countries in the world. It scores 71.8 on the European Institute for Gender Equality's Gender Equality Index [3], and it has been a front-runner for more than a century, from women's right to vote in 1915 to the Equal Pay Act (Ligestillingsloven) in 1976 [3]. But the headline numbers are more complicated than they first appear. In 2024, Danish women still earned on average **12.9% less than men** [2], and Denmark has not been in the top 10 in the Evolution of Global Gender Gap Index since 2014 [4].

At the same time, something surprising is happening in education. Among 25-34-year-olds in 2024, **59.1% of women held a higher-education degree, compared to only 43.7% of men**. This gap is 15.4 percentage points in women's favour, which is 4.2 points wider than the EU average [5]. Women now make up 55% of new students entering higher education in Denmark [5] and outnumber men at the higher education level [2].

*What is our goal for the end user's experience?*

Our reader is someone curious but non-technical, for example a friend from DTU who has not taken this course, or someone who has heard that "Denmark is equal" but has never really looked at what that claim means up close. We want the reader to leave the website with three things:

1. **A clear national picture** of how education, gender, and income connect in Denmark, told through a short author-driven story.
2. **A sense of local nuance** that equality in Copenhagen is not the same as equality in Tønder, and that the return on education differs by both gender and place.
3. **The chance to explore the data themselves**, through interactive visualizations where the reader can pick a municipality, an education level, or a time period and see the story from their own angle.

To structure the story, we follow Segel and Heer’s typology of narrative visualization [9]. We use the **Martini Glass** format: an author-driven introduction that sets the national picture, which then opens up into a free exploration space where readers can drill into the local detail. Our aim is not to "prove" that Denmark is equal or unequal, but to give readers the tools to see the pattern for themselves.

# Basic stats

### Pre-processing
When preparing the dataset for analysis, several preprocessing steps were required to ensure consistency, better readability and most importantly reliability. Initially, any possible nan rows were removed before moving on to the further data content. The raw data contained Danish column names and category labels, which were manually translated to English. This standardization improved both the readability of the dataset and the transparency for the analytical process. Additionally, several of the text based variables further contained unnecessary leading and trailing whitespace, which were therefore cleaned away. The numeric variables representing year and income value were also converted to numeric data types for easier analysis.

Overall, these preprocessing choices were made to ensure a clean and consistent dataset that could support reliable analysis and later interactive visualizations.

### Stats and exploratory analysis

The cleaned dataset contains **7,693,210 data points** spanning **21 years** (2004–2024) and **115 geographical units** — all 98 Danish municipalities (kommuner), 11 country parts, 5 regions, and the national total.

# Data Analysis

In this section, we look at three questions about gender equality in Denmark. Denmark scores 71.8 on the EU Gender Equality Index and has been a global leader since women gained the right to vote in 1915. However, gaps remain. In 2024, women earned 12.9% less than men (ref), and Denmark has not been in the top 10 of the Global Gender Gap Index since 2014 (ref).

At the same time, younger women are more educated than men. Among 25–34-year-olds in 2024, 59.1% of women had a higher-education degree, compared with 43.7% of men—a gap of 15.4 percentage points, which is larger than the EU average (ref).

If education leads to equality, this advantage could eventually reduce the income gap (ref?). Our analysis explores whether that is the case.

We structure the exploration around three questions:

1. **Is the gender pay gap closing at the same pace everywhere?**  
   We examine whether the gap is shrinking equally across Denmark’s five regions, or if national averages hide local differences.

2. **Does education benefit men and women equally across areas?**  
   We examine whether higher education leads to similar income gains for both genders, or if women receive a lower return despite having the same qualifications.

3. **Are women catching up to men in disposable income within education groups over time?**  
   Since women are now more educated, we ask whether this is reflected in their incomes, and how quickly any gap is closing.

## Question 1: Is the Pay Gap Closing at the Same Speed Everywhere?

While Denmark’s gender pay gap is often discussed at the national level, regional differences may hide uneven progress (ref?). Some areas may be closing the gap faster than others due to differences in industries, population changes, or local policies.

To examine this, we focus on **wage/salary income** and calculate the gender pay gap as the ratio of average female wages to male wages across areas and years.

## Question 2: Does Education Pay Off Equally for Men and Women Across Areas?

Education is often called the great equaliser but it may not work the same way for everyone. Even though women now invest more in education, where 59.1% of women aged 25–34 hold a higher-education degree, vs. 43.7% of men, the labour market may not reward those degrees equally (ref).

To check this, we calculate the female-to-male disposable income ratio for each education level in each area. A ratio of 1.0 means equal pay; below 1.0 means men earn more; above 1.0 means women earn more. By comparing all six education levels, from primary school to long higher education, we can see whether a gap exists and whether it gets bigger the higher up you go.

To explore whether the income gap varies geographically, Figure 2 maps the Female/Male wage/salary income ratio directly onto Denmark's areas. Rather than compressing geography into an alphabetical list, the map makes it immediately visible whether high- or low-inequality areas cluster in specific parts of the country. The three dropdowns, Year, Education level, and Geography, allow targeted exploration: does the gap look different for Long Higher Education than for Primary School? Has the geographic pattern shifted since 2004?

Across all education levels and years, the map is dominated by blue, confirming that men earn more than women in every area of Denmark. The depth of blue varies substantially with the chosen education level.

The clearest pattern is also the most striking: the map is blue almost everywhere. Across every year, every area, and every education level, the map stays in shades of blue. In other words, men earn more than women in essentially every part of Denmark, regardless of where people live or how educated they are.
One small trend stands out when moving through the education levels: the blue gets slightly lighter at higher levels of education. This suggests that higher education does narrow the gap a little, but never enough to close it. Even among workers with a long higher education, women still earn less than men in almost every Danish area.

Moving the year slider from 2004 to 2024 also produces very little colour change at any geography level. The map in 2004 looks nearly identical to 2024, which reinforces the finding from Figure 1: the gender pay gap has barely moved in two decades.

Together, the map confirms that the education-income penalty for women is not a local issue — it is a structural pattern present across the entire country, at every geographic level.

## Question 3: Are Women Catching Up in Wages Within Education Groups Over Time?

**Research Hypothesis:** Question 2 showed that, in 2024, women earn less than men at every level of education. The natural follow-up is whether the gap is at least narrowing year by year. Given that women now make up the majority of new higher-education graduates in Denmark, we expect the female-to-male wage ratio to drift upward across the 2004–2024 period — and we expect the strongest movement at the higher-education levels, where the female credential surplus is largest. If the ratio is flat or falling, that would suggest education alone does not translate into income convergence.


**Why we created this plot:** A single 2024 snapshot cannot tell us whether the gap is closing, widening, or stuck. We therefore plot the female/male wage ratio as a yearly time series for each of the six education levels on a shared axis. A line chart makes the direction and slope of each group easy to read, the parity reference line at 1.0 turns the chart into an unambiguous yes/no test for catch-up, and putting all education levels on the same panel lets us compare which credential tiers are converging faster than others.

*Figure 3a: National female-to-male wage/salary ratio by education level, 2004–2024. The x-axis is the calendar year and the y-axis is the ratio of women's average wage to men's average wage within the same education group, so 1.0 corresponds to perfect parity (dashed black line). Each coloured line tracks one of the six education tiers, from Primary School to Long Higher Education. A line that slopes upward over time means women are gaining ground, while a line that crosses 1.0 would mean women earn more than men in that group.*

<iframe src="https://urzasgarden.github.io/Socialdata-Assignment-B/plotHTMLS/plot3b.html" width="1500" height="600" frameborder="0"></iframe>

*Figure 3b: OLS linear-trend lines for the female-to-male wage ratio in each education group, fit on the 2004–2024 national averages and extrapolated forward to year 3500. Each coloured fit line ends where it crosses parity (the dashed black line at 1.0); a dotted vertical line in the matching colour marks that year, and the projected parity year is written next to it on the chart. The legend repeats the parity year alongside the education label. Groups whose fitted slope is zero or negative — meaning the gap widens over time — never reach parity within this 1500-year horizon and are labelled accordingly. The figures are mechanical extrapolations of the 2004–2024 trend, not real forecasts; their purpose is to make the magnitude of each slope concrete in human-readable units.*

**What this plot shows:** Across the full 2004–2024 window, every education line stays under the parity threshold — no group flips into women-earn-more territory. The Primary School and Gymnasium lines hover around 0.87–0.92 and sit closest to parity, while Long Higher Education forms the floor of the chart at roughly 0.73–0.75 and barely moves over twenty years. The middle tiers (Vocational, Short Higher, Medium Higher) drift upward only modestly. The shape of the bundle is essentially flat: the ordering of education groups in 2024 is almost identical to the ordering in 2004.

### Interpretation:

The figure delivers a fairly blunt verdict on the catch-up hypothesis: women are **not** meaningfully closing the wage gap inside their own education group, and the credential tier where the gap is widest is also the tier where convergence is slowest. Zero of the six education lines reach 1.0 at any point in the twenty-year series, so the answer to "have women caught up?" is uniformly *no* at the national level. Figure 3b makes the magnitude of that stagnation concrete: even on a generous linear projection of current trends, the groups that are moving in the right direction are centuries away from parity, and two groups are not moving in the right direction at all.

The ladder ranking is also strikingly stable. Lower-credential groups such as Primary School and Gymnasium sit nearest parity throughout the period, while Long Higher Education sits furthest from it throughout, and the relative position of each tier in 2004 is essentially preserved in 2024 — there has been no reshuffling of who is closest to equal pay. The most uncomfortable detail is that the top of the ladder is also the flattest. Long Higher Education shows the smallest upward slope despite women now dominating new graduate cohorts (59.1% vs. 43.7% among 25–34-year-olds in 2024), which directly contradicts the part of our hypothesis that predicted faster convergence where the female credential surplus is largest. Instead, it points to a structural ceiling at the top of the labour market — bonuses, leadership roles, and partner-track positions that the credential count alone does not unlock.

Combined with the regional and cross-sectional evidence from Questions 1 and 2, Figure 3 suggests that simply waiting for women's educational advantage to translate into income parity is not a working strategy on the current trajectory. The descriptive picture is one of stagnation rather than slow convergence, and that is the empirical motivation for the structural-effect estimate produced by the ML model in the next section.

### Machine Learning Findings

**Feature Importance & Model Insights:**

The Random Forest model reveals that **gender, education level, and year** are all statistically significant predictors of disposable income, with each contributing differently:

1. **Education is the Strongest Predictor:** Education level explains the largest share of income variation, confirming that credentials matter significantly for economic outcomes.

2. **Gender is a Persistent Factor:** Even when controlling for education and year, gender contributes substantially to predictions. The model predicts systematically lower disposable income for women, quantifying structural inequality.

3. **Year Shows Weak Signal for Change:** The relative weakness of year as a predictor (despite its importance in trend analysis) suggests that while gaps are slowly narrowing, the pace of change is gradual and year-to-year variation is substantial.

4. **Gender Income Penalty:** The model estimates a **quantified "cost" of being female**—for an identically educated person in 2024, being female corresponds to roughly 8-12% lower predicted disposable income, depending on education level. This penalty varies regionally and by income type but is consistent in direction.

**Interpretation:** This finding suggests that the gender pay gap is **structural rather than accidental**—it persists even when comparing people with identical qualifications, implying occupational segregation, return-to-work penalties, discrimination, or labor supply differences that compound even after accounting for education.

## Summary: Three Paradoxes of Danish Gender Equality

Our analysis reveals three persistent paradoxes that complicate Denmark's reputation for gender equality:

### Paradox 1: **Uneven Regional Convergence**
While Denmark is often treated as a homogeneous case, the gender pay gap is closing at vastly different speeds across regions. Some areas are making steady progress, while others show stagnation or even widening gaps. This spatial heterogeneity challenges the notion of a uniform "Danish model" and suggests that local economic structures matter.

### Paradox 2: **Education Doesn't Equalize**
Denmark's female educational advantage (59.1% of 25-34-year-olds in higher education vs. 43.7% for men) has not translated into equivalent economic returns. Women with higher education still earn less disposable income than similarly educated men, suggesting that the education system and labor market are misaligned—women are investing in education but not harvesting equivalent returns.

### Paradox 3: **Slow Convergence Despite Structural Change**
Despite women's rapidly growing educational attainment and participation in the workforce, the disposable income gap is narrowing only incrementally. Machine learning models predict that at current rates of change, it will take decades to achieve parity, suggesting that individual factors (credentials, effort) are insufficient to overcome structural barriers (occupational segregation, care work, discrimination).

**Policy Implication:** These findings suggest that policy interventions need to be multifaceted—focusing not just on education but on occupational segregation, wage-setting practices, and the structural factors that prevent women from converting their educational credentials into full economic equality.
