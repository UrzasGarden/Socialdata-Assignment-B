# Motivation

Our primary dataset comes from Statistics Denmark (Danmarks Statistik), the official authority responsible for producing statistics on Danish society [1]. We work with a merged dataset drawn from StatBank, combining official registers on income, education, and gender across Danish municipalities (kommuner) from 2004 to 2024.

The full dataset contains 7,693,210 data points across 7 variables:

1. **Year** - covering 21 years (2004-2024), which allows us to study change over time.
2. **Area** - 115 geographical units, including municipalities and regions, so we can make local comparisons.
3. **Gender** - Male, Female, and Both, which makes direct gender-gap analysis possible.
4. **Education level** - 7 categories, ranging from Primary Education to Long Higher Education, plus Not Stated.
5. **Income type** - 39 categories, such as Disposable income, Pay, Pension, and Transfer payments.
6. **Value** - the measured quantity, which ranges from about -1.8 million to 600 million.
7. **Unit** - 4 measurement units, for example income amounts in 1,000 DKK or number of persons.

*Why did we choose this dataset?*

Denmark is often described as one of the most gender-equal countries in the world. It scores 71.8 on the European Institute for Gender Equality's Gender Equality Index [3], and it has been a front-runner for more than a century, from women's right to vote in 1915 to the Equal Pay Act (Ligestillingsloven) in 1976 [3]. But the headline numbers are more complicated than they first appear. In 2024, Danish women still earned on average **12.9% less than men** [2], and Denmark has not been in the top 10 in the Evolution of Global Gender Gap Index since 2014 [4].

At the same time, something surprising is happening in education. Among 25-34-year-olds in 2024, **59.1% of women held a higher-education degree, compared to only 43.7% of men**. This gap is 15.4 percentage points in women's favour, which is 4.2 points wider than the EU average [5]. Women now make up 55% of new students entering higher education in Denmark [5] and outnumber men at the higher education level [2].

*What is our goal for the end user's experience?*

Our reader is someone curious but non-technical, for example a friend from DTU who has not taken this course, or someone who has heard that "Denmark is equal" but has never really looked at what that claim means up close. We want the reader to leave the website with three things:

1. **A clear national picture** of how education, gender, and income connect in Denmark, told through a short author-driven story.
2. **A sense of local nuance** that equality in Copenhagen is not the same as equality in Tønder, and that the return on education differs by both gender and place.
3. **The chance to explore the data themselves**, through interactive visualizations where the reader can pick a municipality, an education level, or a time period and see the story from their own angle.

To structure the story, we follow Segel and Heer’s typology of narrative visualization [9]. We use the **Martini Glass** format: an author-driven introduction that sets the national picture, which then opens up into a free exploration space where readers can drill into the local detail. Our aim is not to "prove" that Denmark is equal or unequal, but to give readers the tools to see the pattern for themselves.

# Basic stats

### Pre-processing
When preparing the dataset for analysis, several preprocessing steps were required to ensure consistency, better readability and most importantly reliability. Initially, any possible nan rows were removed before moving on to the further data content. The raw data contained Danish column names and category labels, which were manually translated to English. This standardization improved both the readability of the dataset and the transparency of the analytical process. Additionally, several of the text based variables further contained unnecessary leading and trailing whitespace, which were therefore cleaned away. The numeric variables representing year and income value were also converted to numeric data types for easier analysis.

Overall, these preprocessing choices were made to ensure a clean and consistent dataset that could support reliable analysis and later interactive visualizations.

### Stats and exploratory analysis

The cleaned dataset contains **7,693,210 data points** spanning **21 years** (2004–2024) and **115 geographical units** — all 98 Danish municipalities (kommuner), 11 landsdele, 5 regions, and the national total.

**Temporal coverage** is mostly consistent throughout, however with a few jumps in data collection. The year distribution bar chart below shows each of the 21 years is represented with roughly 345 - 377 thousand data points per year. This range is not dramatic, making the dataset well suited for studying long-term trends without worrying about larger collection irregularities. The visible increase in data collection in 2010 is due to a regulation change, where imigrants were allowed to registrer their educations obtained in other countries than Denmark. The next, smaller, jump in 2017 is explained by the old publication being discontinued, and a greater emphasis was now placed on online dissemination, creating more comparable statistics on incomes and therefore slightly more data points. [10]




**Income type coverage** is similarly largely stable. All **39 income types** - from wages and business profits to public pensions and disposable income - are tracked consistently across years with roughly 201-201.8 thousand data points per category. The income type coverage bar chart below Clarifies the distribution with a number color coding (ligher color = fewer data points). Minor coverage gaps exist for three specific income types: "Green Tax Compensation", "Property tax (land tax), homeowners", and "Property tax (land tax), tenants". These categories arrived later that the remaining and therefore contain noticably fewer data points. It should also be mentioned that items such as lottery winnings and undeclared income are not covered by the statistics. [10]




**Education-level coverage** across all seven levels is visible in the stacked area chart below. It covers Unknown and Primary School through Long Higher Education. It is visible how the distribution of categories are exactly equal across the years.



**Gender representation** is perfectly balanced by design. The stacked bar chart below of gender distribution by year shows Male, Female, and Both Genders Total each accounting for exactly one-third of the data in every year. This is because Statistics Denmark records each income statistic separately for all three gender categories for every area–education–income-type combination, so the split is structural rather than coincidental.



**Geographic coverage** was validated through choropleth maps at three levels of geography. At the municipality level, 97 out of 99 municipalities in the boundary file matched to data entries, and all 11 country parts and all 5 regions were fully matched.
